# Import Module

```python
from tensorflow.keras.models import Sequential #model creating
from tensorflow.keras.layers import Flatten # converting in one 1D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Conv2D #convolutional layer creating
from tensorflow.keras.layers import MaxPooling2D
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator # data loading
```

# Step 1
## collect the data

##  example of progressively loading images from file
### create generator 

```python
datagen = ImageDataGenerator()
```

#### prepare an iterators for each dataset

```python
train_it = datagen.flow_from_directory('train',
                                       class_mode='binary',
                                       batch_size=64,
                                       target_size=(64,64)) #target size by default(256,256)
```

```python
val_it = datagen.flow_from_directory('test',
                                     class_mode='binary',
                                     batch_size=64,
                                     target_size=(64,64))
```

#### confirm the iterator works

```python
batchX, batchy = train_it.next()
print('Batch shape=%s, min=%.3f, max=%.3f' % (batchX.shape, batchX.min(), batchX.max()))
```

# Step 2
## Build model

```python
classifier = Sequential()

classifier.add(Conv2D(32,(3,3),input_shape=(64,64,3),activation = 'relu'))
classifier.add(MaxPooling2D(pool_size=(2,2),strides=2)) #if stride not given it equal to pool filter size

classifier.add(Conv2D(32,(3,3),activation = 'relu'))
classifier.add(MaxPooling2D(pool_size=(2,2),strides=2))

classifier.add(Flatten()) #it convert 2D to 1d form

classifier.add(Dense(units=128,activation='relu'))
classifier.add(Dense(units=1,activation='sigmoid'))


```

### Configures the model for training.

```python
classifier.compile(optimizer='rmsprop',
                   loss='binary_crossentropy',
                   metrics=['accuracy'])
```

# Step 3
## Train the model

```python
classifier.fit_generator(train_it,
                        steps_per_epoch=10, 
                        epochs = 1250,
                        validation_data = val_it,
                        validation_steps = 20)
classifier.save_weights('models/simple_CNN.h5')
```
